Section 8: End-to-end flow

```python
print("=" * 60)
print("SECTION 8: End-to-end flow")
print("=" * 60)
 
candidates = find_product_by_query.invoke({"query": "building blocks"})
top_match = candidates[0]
print(f"Step 1 - top match: {top_match['name']} (id={top_match['id']})")
print()
 
info = get_product_info.invoke({"product_id": top_match["id"]})
print(f"Step 2 - full info:")
for k, v in info.items():
    print(f"  {k}: {v}")
print()
 
stock = check_inventory.invoke({"product_id": top_match["id"]})
print(f"Step 3 - stock:")
for k, v in stock.items():
    print(f"  {k}: {v}")
print()
 
quote = calculate_quote.invoke({"product_id": top_match["id"], "quantity": 60})
print(f"Step 4 - quote:")
for k, v in quote.items():
    print(f"  {k}: {v}")
```
